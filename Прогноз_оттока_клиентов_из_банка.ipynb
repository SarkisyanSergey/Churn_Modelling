# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

```python
def plot_roc():
    """Рисует график ROC кривой"""
    plt.figure(figsize=(10, 6))
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.plot([0, 1], [0, 1], linestyle='--', label = 'Dummy: 0.5')
```

```python
STATE = 12345
scoring = ['f1','roc_auc', 'recall']
#Таблица для результатов экспериментов
index = []
scores = {'dataset': [],
          'f1': [],
          'roc_auc': [],
          'recall': [],
          'best_score_': [],
          'best_params_': [],
         }
def result_append(model_name, dataset, y_test, y_pred, gridsearch):
    """добавляет данные в таблицу экспериментов и рисует график"""
    index.append(model_name)
    scores['dataset'].append(dataset)
    scores["f1"].append(f1_score(y_test, y_pred))
    scores["roc_auc"].append(roc_auc_score(y_test, y_pred))
    scores["recall"].append(recall_score(y_test, y_pred))
    scores["best_score_"].append(gridsearch.best_score_)
    scores["best_params_"].append(gridsearch.best_params_)
    return pd.DataFrame(scores, index=index)
```

```python
data = df.copy(deep = True)
target = data['Exited']
features = data.drop('Exited', axis=1)
```

```python
#Разделение данных
x, x_test, y, y_test = train_test_split(
    features, target, test_size=0.2, stratify=target, random_state=STATE)

x_train, x_valid, y_train, y_valid = train_test_split(
    x, y, test_size=0.25, stratify = y, random_state=STATE)
```

```python
print('Размер валидационной выборки: ', x_valid.shape )
print('Размер тестовой выборки: ', x_test.shape )
```

```python
class GetDummies(BaseEstimator, TransformerMixin):
    def __init__(self, drop_first=False):
        self.drop_first = drop_first
    def fit(self, x, y=None):
        return self
    def transform(self, x):
        return pd.get_dummies(x, drop_first=self.drop_first)
```

```python
num_pipe = make_pipeline(StandardScaler())
cat_pipe = make_pipeline(GetDummies(drop_first=True))

preproc = make_column_transformer(
    (num_pipe, selector(dtype_include="number")),
    (cat_pipe, selector(dtype_include="category")),
)
```

```python
LR_opt_params = {
    'LogisticRegression__penalty': ['l1','l2'],
    'LogisticRegression__C': [0.01, 0.1, 1, 10, 20]
}

SP = {'LogisticRegression':LR_opt_params,

     }
```

```python
cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=5, random_state=STATE)
```

```python
models = [
    ['LogisticRegression', LogisticRegression(random_state=STATE)],
    
     ]
```

```python
%%time
plot_roc()

for model in models:
    pipeline = Pipeline(steps=[['preproc', preproc], model])
    gridsearch = RandomizedSearchCV(estimator=pipeline,
                                    param_distributions=SP[model[0]],
                                    scoring='f1',
                                    cv=cv)
    gridsearch.fit(x_train, y_train)
    y_pred = gridsearch.predict(x_valid)
    result_append(model[0], 'Imbalanced', y_valid, y_pred, gridsearch)
    probabilities_valid = gridsearch.predict_proba(x_valid)
    probabilities_one_valid = probabilities_valid[:, 1]
    fpr, tpr, thresholds = roc_curve(y_valid, probabilities_one_valid)
    label = f'{model[0]}: {round(roc_auc_score(y_valid, probabilities_one_valid),2)}'
    plt.plot(fpr, tpr, label=label)

plt.legend()
plt.show()
pd.DataFrame(scores, index=index)
```
